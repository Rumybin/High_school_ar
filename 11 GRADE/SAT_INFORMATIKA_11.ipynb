**Membuat kolom baru bernama "month" yang menggelompokkan date bulan **

```python
Data3['month'] = Data3['Date'].dt.month_name()
```

```python
Data3
```

**Membuat Group yang terisi month dan New cases**

```python
monthly_cases = (
    Data3.groupby('month')['New Cases']
    .sum()
    .reset_index(name='Total New Cases')
)
```

**Dengan format bulan 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'**

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_cases['month'] = pd.Categorical(monthly_cases['month'], categories=month_order, ordered=True, )
monthly_cases = monthly_cases.sort_values('month')
```

**Nilai month yang telah di group di sort urut berdasarkan urutan input = inplace true**

```python
Data3.sort_values('month', inplace=True)
```

```python
print(monthly_cases)
```

Menampilkan data menggunakan px.line sehingga hanya 1 garis pada diagram untuk membuat hubungan antara x='month', y='Total New Cases'

```python
px.line(monthly_cases, x='month', y='Total New Cases', title='Covid end')
```
